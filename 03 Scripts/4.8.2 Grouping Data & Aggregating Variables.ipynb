#### Creating a Loyalty flag based on number of orders per user

```python
#Create a 'max_order' variable in ords_prods_merge to signify the max order per user_id
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Checking results
ords_prods_merge.head(20)
```

```python
# Making Loyal customer val in 'loyalty_flag' col based on transformation results
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# Making Regular customer val in 'loyalty_flag' col based on transformation results
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# Making New customer val in 'loyalty_flag' col based on transformation results
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

### 6. Create a spending flag for each user based on the _mean_ price across all their orders.
Criteria:
</br>High spender >= 10
</br>Low spender < 10

```python
#Create a 'mean_price' variable to signify the mean price per user_id
ords_prods_merge['mean_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(lambda s: s.loc[[True if (v != 99999) & (v != 14900) else False for v in s]].mean())
```

```python
#Checking results
ords_prods_merge.shape
```

```python
#Checking results
ords_prods_merge[['user_id', 'prices', 'mean_price']].head(50)
```

```python
#Checking 'mean_price' min, max, and mean
ords_prods_merge['mean_price'].min(),ords_prods_merge['mean_price'].max(), ords_prods_merge['mean_price'].mean()
```

```python
# Making 'High spender' val in 'spender_flag' col based on transformation results
ords_prods_merge.loc[ords_prods_merge['mean_price'] >= 10, 'spender_flag'] = 'High spender'
```

```python
# Making 'Low spender' val in 'spender_flag' col based on transformation results
ords_prods_merge.loc[ords_prods_merge['mean_price'] < 10, 'spender_flag'] = 'Low spender'
```

```python
#Checking results
ords_prods_merge.shape
```

```python
#Checking results
ords_prods_merge[['user_id', 'mean_price', 'spender_flag']].tail(60)
```

### 7. Create an order frequency flag that marks the regularity of a users ordering behavior according to the _median_ in the days_since_prior_order column.
Criteria:
</br>Non-frequent customer > 20
</br>Regular customer <= 20 & > 10
</br>Frequent customer <= 10

```python
#Create a 'median_order_frequency' variable to signify the median 'days_since_prior_order' per user_id
#ords_prods_2['median_order_frequency'] = ords_prods_2.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
ords_prods_merge['median_order_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(lambda s: s.loc[[True if v != 999 else False for v in s]].median())
```

```python
#Checking results
ords_prods_merge.shape
```

```python
#Checking results
ords_prods_merge[['user_id', 'days_since_prior_order', 'median_order_frequency']].head()
```

```python
#Checking 'median_order_frequency' min, max, and mean
ords_prods_merge['median_order_frequency'].min(),ords_prods_merge['median_order_frequency'].max(), ords_prods_merge['median_order_frequency'].mean()
```

```python
# Making 'Non-frequent customer' val in 'order_frequency_flag' col based on transformation results
ords_prods_merge.loc[ords_prods_merge['median_order_frequency'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
# Making 'Regular customer' val in 'order_frequency_flag' col based on transformation results
ords_prods_merge.loc[(ords_prods_merge['median_order_frequency'] <= 20) & (ords_prods_merge['median_order_frequency'] > 10), 'order_frequency_flag'] = 'Regular customer'
```

```python
# Making 'Frequent customer' val in 'order_frequency_flag' col based on transformation results
ords_prods_merge.loc[ords_prods_merge['median_order_frequency'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
#Checking results
ords_prods_merge.shape
```
